```python
# Q1. Discuss string slicing and provide examples.




# String slicing allows you to extract a part of a string by specifying a start position, an end position, and an optional step value. Think of a string as a list of characters, and slicing helps you pick a subset of those characters.

#Basic Syntax

# The general syntax for slicing a string is:

string [start:end:step]

# - start : The index where the slice starts (inclusive). If omitted, it defaults to the beginning of the string.
# - end : The index where the slice ends (exclusive). If omitted, it defaults to the end of the string.
# - step : The step value determines the stride between elements. If omitted, it defaults to 1.

# Examples :

# 1. Basic Slicing :

text = "Hello, World!"
slice1 = text[0:5] # Extracts characters from index 0 to 4
    print(slice1)  # Output: Hello


2. Omitting start and end

text = "Hello, World!"
slice2 = text[:5]  # Omits start, so it starts from the beginning
    print(slice2)  # Output: Hello
slice3 = text[7:]  # Omits end, so it goes to the end
    print(slice3)  # Output: World!


# 3. Using step : 

text = "Hello, World!"
slice4 = text[::2]  # Takes every second character
    print(slice4)  # Output: Hlo ol!

# 4.Negative Indexes :

text = "Hello, World!"
slice5 = text[-6:-1]  # Negative indexing counts from the end
    print(slice5)  # Output: World


# 5. Reversing a String :

text = "Hello, World!"
reversed_text = text[::-1]  # Reverses the entire string
    print(reversed_text)  # Output: !dlroW ,olleH
```

```python
# Q2. Explain the key features of lists in Python.



# Lists in Python are versatile, ordered collections of items. They are one of the most commonly used data structures in Python due to their flexibility and ease of use. Here are some of the key features of lists:

# 1. Ordered Collection

# Definition : Lists maintain the order of elements. This means that items have a specific sequence, and the order in which you add items is preserved.
# Example :

fruits = ['apple', 'banana', 'cherry']
    print(fruits)  # Output: ['apple', 'banana', 'cherry']


# 2. Mutable : 

# Definition : Lists are mutable, which means you can change their contents (add, remove, or modify elements) after they are created.
# Example :

fruits = ['apple', 'banana', 'cherry']
fruits[1] = 'blueberry'  # Modify the second item
    print(fruits)  # Output: ['apple', 'blueberry', 'cherry']


# 3. Dynamic Size :

# Definition : Lists can grow or shrink in size. You can add or remove elements as needed.
# Example :

fruits = ['apple', 'banana']
fruits.append('cherry')  # Add an item
    print(fruits)  # Output: ['apple', 'banana', 'cherry']
fruits.pop()  # Remove the last item
    print(fruits)  # Output: ['apple', 'banana']

#   
# 4. Allow Duplicate Elements :

# Definition : Lists can contain duplicate items. Each element is stored with its own index.
# Example :

numbers = [1, 2, 2, 3, 4, 4]
    print(numbers)  # Output: [1, 2, 2, 3, 4, 4]


# 5. Can Contain Mixed Data Types :

# Definition : Lists can hold items of different data types within the same list.
# Example :

mixed_list = [1, 'apple', 3.14, True]
    print(mixed_list)  # Output: [1, 'apple', 3.14, True]


# 6. Indexing and Slicing :

# Definition : You can access elements of a list using zero-based indexing. Slicing allows you to get a subset of elements.
# Example :

fruits = ['apple', 'banana', 'cherry']
    print(fruits[0])  # Output: 'apple'
    print(fruits[1:3])  # Output: ['banana', 'cherry']


# 7. Supports Nested Lists :

# Definition : Lists can contain other lists, allowing for multi-dimensional structures.
# Example :

nested_list = [[1, 2], [3, 4], [5, 6]]
    print(nested_list[1])  # Output: [3, 4]
    print(nested_list[1][0])  # Output: 3


# 8. Built-in Methods :

# Definition : Lists come with a variety of built-in methods for common operations, such as adding, removing, or sorting elements.
# Example :

fruits = ['apple', 'banana', 'cherry']
fruits.append('date')  # Add an item
fruits.remove('banana')  # Remove an item
fruits.sort()  # Sort the list
    print(fruits)  # Output: ['apple', 'cherry', 'date']
  
# 

# 9. Iteration :

# Definition : Lists can be easily iterated over using loops.
# Example :

fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
      print(fruit)
  # Output:
  # apple
  # banana
  # cherry


# 10. List Comprehensions :

# Definition : Python supports a concise way to create lists based on existing lists using list comprehensions.
# Example :

numbers = [1, 2, 3, 4, 5]
squares = [x**2 for x in numbers]
  print(squares)  # Output: [1, 4, 9, 16, 25]
  
# 
# Lists in Python are incredibly flexible and useful, making them one of the most important data structures to understand for effective programming.
```

```python
# Q3. Describe how to access, modify, and delete elements in a list with examples.




# Accessing Elements

# You can access elements in a list using indexing. Python uses zero-based indexing, so the first element has an index of 0, the second has an index of 1, and so on.

# Examples:

# 1. Accessing a Single Element :

fruits = ['apple', 'banana', 'cherry']
   print(fruits[0])  # Output: 'apple' (first element)
print(fruits[2])  # Output: 'cherry' (third element)
   

# 2. Accessing with Negative Indexes :

print(fruits[-1])  # Output: 'cherry' (last element)
   print(fruits[-2])  # Output: 'banana' (second-to-last element)


# 3. Slicing to Access Multiple Elements :

print(fruits[1:3])  # Output: ['banana', 'cherry'] (elements from index 1 to 2)
print(fruits[:2])   # Output: ['apple', 'banana'] (elements from the start up to index 2)
   

# Modifying Elements :
 
# Lists are mutable, meaning you can change their content. You can modify elements using indexing or through methods.

# Examples :

# 1. Modifying an Element :

fruits = ['apple', 'banana', 'cherry']
fruits[1] = 'blueberry'  # Change the second element
   print(fruits)  # Output: ['apple', 'blueberry', 'cherry']


# 2. Modifying Multiple Elements with Slicing :

fruits = ['apple', 'banana', 'cherry']
fruits[1:3] = ['blueberry', 'date']  # Replace elements from index 1 to 2
   print(fruits)  # Output: ['apple', 'blueberry', 'date']


# 3. Appending and Inserting Elements :

fruits = ['apple', 'banana']
fruits.append('cherry')  # Add 'cherry' at the end
   print(fruits)  # Output: ['apple', 'banana', 'cherry']
fruits.insert(1, 'blueberry')  # Insert 'blueberry' at index 1
print(fruits)  # Output: ['apple', 'blueberry', 'banana', 'cherry']
   

# Deleting Elements :

# You can remove elements from a list using several methods, including `del`, `pop()`, and `remove()`.

# Examples:

# 1. Using `del` :

fruits = ['apple', 'banana', 'cherry']
del fruits[1]  # Remove the element at index 1
   print(fruits)  # Output: ['apple', 'cherry']


# 2. Using `pop()` :

fruits = ['apple', 'banana', 'cherry']
removed_item = fruits.pop()  # Remove and return the last element
   print(removed_item)  # Output: 'cherry'
print(fruits)  # Output: ['apple', 'banana']
removed_item = fruits.pop(0)  # Remove and return the element at index 0
   print(removed_item)  # Output: 'apple'
print(fruits)  # Output: ['banana']
   

# 3. Using `remove()` :

fruits = ['apple', 'banana', 'cherry']
fruits.remove('banana')  # Remove the first occurrence of 'banana'
   print(fruits)  # Output: ['apple', 'cherry']


# 4. Clearing the Entire List :

fruits = ['apple', 'banana', 'cherry']
fruits.clear()  # Remove all elements
    print(fruits)  # Output: []
    
# Summary

# - Accessing : Use indexing (`list[index]`) or slicing (`list[start:end]`).
# - Modifying : Change elements directly using indexing or use methods like `append()`, `insert()`, and slicing.
# - Deleting : Use `del` to remove by index, `pop()` to remove and get an element, `remove()` to remove by value, and `clear()` to empty the list.

# These operations allow you to effectively manage and manipulate lists in Python.
```

```python
# Q5. Describe the key features of sets and provide examples of their use.



# Sets in Python are a built-in data type used to store collections of unique items. They are similar to mathematical sets and offer several key features and operations. Here’s a breakdown of their key features along with examples:

# Key Features of Sets

# 1. Unordered Collection :

# Definition : Sets do not maintain any specific order of elements. When you iterate over a set, the order of elements is not guaranteed.
# Example :
fruits = {'apple', 'banana', 'cherry'}
    print(fruits)  # Output: {'apple', 'banana', 'cherry'} (order is not guaranteed)


# 2. Unique Elements :
# Definition : Sets automatically eliminate duplicate values. Each element must be unique.
# Example :
fruits = {'apple', 'banana', 'apple', 'cherry'}
    print(fruits)  # Output: {'apple', 'banana', 'cherry'} (duplicates are removed)


# 3. Mutable :
# Definition : Sets are mutable, meaning you can add or remove elements after the set is created.
# Example :
fruits = {'apple', 'banana'}
fruits.add('cherry')  # Add a new item
    print(fruits)  # Output: {'apple', 'banana', 'cherry'}
fruits.remove('banana')  # Remove an item
    print(fruits)  # Output: {'apple', 'cherry'}



# 4. No Indexing :
# Definition : Since sets are unordered, you cannot access elements by index. You can only iterate over them.
# Example :

fruits = {'apple', 'banana', 'cherry'}
    print(fruits[1])  # This will raise an error
for fruit in fruits:
    print(fruit)  # Output: each fruit in no particular order
     

# 5. Supports Mathematical Operations :

# Definition : Sets support operations like union, intersection, difference, and symmetric difference.
# Examples :
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
     
# Union
union_set = set1 | set2
    print(union_set)  # Output: {1, 2, 3, 4, 5, 6}
     
# Intersection
intersection_set = set1 & set2
    print(intersection_set)  # Output: {3, 4}
     
# Difference
difference_set = set1 - set2
    print(difference_set)  # Output: {1, 2}
     
# Symmetric Difference
sym_diff_set = set1 ^ set2
    print(sym_diff_set)  # Output: {1, 2, 5, 6}

# 6. Set Comprehensions :
# Definition : Similar to list comprehensions, sets can be created using set comprehensions.
# Example :

squares = {x**2 for x in range(5)}
    print(squares)  # Output: {0, 1, 4, 9, 16}

# 7. Frozensets :

# Definition : Frozensets are immutable versions of sets. They cannot be modified after creation.
# Example :
frozen_fruits = frozenset({'apple', 'banana', 'cherry'})
    print(frozen_fruits)  # Output: frozenset({'apple', 'banana', 'cherry'})

frozen_fruits.add('date')  # This will raise an error: AttributeError

# Use Cases :

# 1. Removing Duplicates :
# Example : Convert a list with duplicates into a set to remove duplicates.
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = set(numbers)
    print(unique_numbers)  # Output: {1, 2, 3, 4, 5}

# 2. Membership Testing :
# Example : Efficiently check if an item is present in a set.

fruits = {'apple', 'banana', 'cherry'}
    print('banana' in fruits)  # Output: True
    print('grape' in fruits)  # Output: False

# 3. Set Operations in Mathematical and Logical Problems :
# Example : Use set operations to find common elements between two lists.

list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
common_elements = set(list1) & set(list2)
    print(common_elements)  # Output: {3, 4}
    

# 4. Efficient Removal of Duplicate Entries :
# Example : Use sets to filter out duplicates from a collection of data.
data = ['apple', 'banana', 'cherry', 'apple', 'banana']
unique_data = set(data)
    print(unique_data)  # Output: {'apple', 'banana', 'cherry'}

# Sets are a powerful tool in Python for handling unique collections and performing mathematical set operations efficiently.
```

```python
# Q6. Discuss the use cases of tuples and sets in Python programming.



# Tuples and sets are both fundamental data structures in Python, and they serve different purposes. Here’s a detailed look at their use cases:

# Tuples :

# 1. Immutable Collections : Tuples are immutable, meaning their contents cannot be changed once they are created. This immutability is useful when you want to ensure that a collection of items remains constant throughout the program. For example, you might use tuples to represent fixed configurations or records that should not be modified.

# 2. Multiple Return Values : Functions in Python can return multiple values as a tuple. This is a common pattern for returning multiple pieces of related information from a function. For instance:
def get_min_max(numbers):
    return (min(numbers), max(numbers))

# 3. Dictionary Keys : Since tuples are immutable, they can be used as keys in dictionaries. This is not possible with lists, which are mutable. For example, you might use a tuple to represent a coordinate in a grid:
grid = {}
grid[(1, 2)] = "value"


# 4. Data Records : Tuples can be used to group related data together. This is particularly useful in cases where you want to return a record-like structure without defining a custom class. For instance:
person = ("John", "Doe", 30)


# 5. Iteration : Tuples can be used in iteration to unpack values conveniently. For example, when iterating over a list of tuples:
for name, age in [("Alice", 30), ("Bob", 25)]:
    print(name, age)

# Sets :

# 1. Unique Elements : Sets are used to store unique elements. They automatically remove duplicates, which is useful when you need to maintain a collection of distinct items. For instance:
unique_numbers = set([1, 2, 2, 3, 4])
   # unique_numbers will be {1, 2, 3, 4}

# 2. Membership Testing : Sets provide an efficient way to test membership. Checking if an item is in a set is generally faster than checking if an item is in a list. For example:
unique_set = {1, 2, 3}
if 2 in unique_set:
    print("2 is in the set")

# 3. Set Operations : Sets support mathematical operations like union, intersection, difference, and symmetric difference. These operations are useful in scenarios where you need to compare and combine collections of data. For example:
set1 = {1, 2, 3}
set2 = {3, 4, 5}
union = set1 | set2        # {1, 2, 3, 4, 5}
intersection = set1 & set2 # {3}
difference = set1 - set2   # {1, 2}

# 4. Removing Duplicates : Sets are useful for removing duplicates from a collection. If you have a list with duplicate items and you need a list with only unique items, you can convert it to a set and back to a list:
list_with_duplicates = [1, 2, 2, 3, 4, 4]
unique_list = list(set(list_with_duplicates))

# 5. Performance Considerations : Sets are implemented using hash tables, which generally offer faster membership testing and element addition/removal compared to lists, especially for large datasets.

# In summary:
# Tuples are great for fixed collections of items and data that should not change, while sets are ideal for collections of unique elements and efficient membership testing. Each data structure has its own strengths and is suited for different types of tasks in Python programming.
```

```python
# Q7. Describe how to add, modify, and delete items in a dictionary with examples.


# Dictionaries in Python are versatile data structures that store key-value pairs. They are mutable, meaning you can add, modify, and delete items dynamically. Here’s how you can perform these operations with examples:

# 1. Adding Items :

# To add a new item to a dictionary, you simply assign a value to a new key. If the key doesn’t exist, it will be created along with its value.

# Example :
# Initialize a dictionary
person = {'name': 'Alice', 'age': 30}

# Add a new key-value pair
person['email'] = 'alice@example.com'

print(person)
# Output: {'name': 'Alice', 'age': 30, 'email': 'alice@example.com'}


# You can also use the `update()` method to add multiple key-value pairs at once:

**Example:**
# Initialize a dictionary
person = {'name': 'Bob', 'age': 25}

# Add multiple key-value pairs
person.update({'email': 'bob@example.com', 'city': 'New York'})

print(person)
# Output: {'name': 'Bob', 'age': 25, 'email': 'bob@example.com', 'city': 'New York'}


# 2. Modifying Items :

# To modify an existing item, you assign a new value to an existing key. The key must already be in the dictionary; otherwise, a new key-value pair will be added.

# Example :
# Initialize a dictionary
person = {'name': 'Charlie', 'age': 35}

# Modify an existing key's value
person['age'] = 36

print(person)
# Output: {'name': 'Charlie', 'age': 36}


# You can also use the `update()` method to modify one or more items:

# Example :
# Initialize a dictionary
person = {'name': 'Diana','age': 40, 'email': 'diana@example.com'}

# Modify existing items
person.update({'age': 41, 'email': 'diana_new@example.com'})

print(person)
# Output: {'name': 'Diana', 'age': 41, 'email': 'diana_new@example.com'}

# 3. Deleting Items :

# To delete an item, you can use the `del` statement or the `pop()` method. The `pop()` method also allows you to handle cases where the key might not exist by providing a default value.

# Example with `del`:
# Initialize a dictionary
person = {'name': 'Eve', 'age': 45, 'email': 'eve@example.com'}
del person['email']
print(person)
# Output: {'name': 'Eve', 'age': 45}

# Example with `pop()`:
# Initialize a dictionary
person = {'name': 'Frank', 'age': 50, 'email': 'frank@example.com'}
email = person.pop('email')
print(email)    # Output: frank@example.com
print(person)   # Output: {'name': 'Frank', 'age': 50}

# Delete a key-value pair with a default value if the key doesn't exist
phone = person.pop('phone', 'Not available')
print(phone)    # Output: Not available

# Summary

# Adding Items : Use `dict[key] = value` or `dict.update({key: value})`.
# Modifying Items : Use `dict[key] = new_value` or `dict.update({key: new_value})`.
# Deleting Items : Use `del dict[key]` or `dict.pop(key, default_value)`.

# These operations allow you to manage and manipulate dictionaries effectively in Python.
```

```python
# Q8. Discuss the importance of dictionary keys being immutable and provide examples.



# Dictionary keys in Python need to be immutable (unchangeable) for a few important reasons. Here's a simple explanation:

# Why Dictionary Keys Need to Be Immutable :

# 1. Stable Lookup : Dictionaries use a system called hashing to quickly find values associated with keys. For this system to work properly, the key must not change. If the key were allowed to change, the dictionary wouldn’t be able to find it efficiently, leading to errors or slower lookups.

# 2. Consistency : Immutable keys ensure that once a key is set, it remains the same. This consistency helps keep the dictionary reliable, as the key will always point to the correct value.

# 3. Efficiency : With immutable keys, the dictionary can perform operations like searching, adding, and removing items quickly and effectively. Changing keys could disrupt these operations.

# Examples of Immutable and Mutable Keys :

# Immutable Keys :

# Immutable keys include types like strings, numbers, and tuples (as long as the tuples contain only other immutable types). These types can be used as dictionary keys.

# Example 1: Using Strings as Keys :
# Using a string as a key
person = {'name': 'Alice', 'age': 30}
    print(person['name'])  # Output: Alice


# Example 2: Using Tuples as Keys :
# Using a tuple as a key
locations = {(37.7749, -122.4194): 'San Francisco', (34.0522, -118.2437): 'Los Angeles'}
    print(locations[(37.7749, -122.4194)])  # Output: San Francisco

# Mutable Keys :

# Mutable keys, like lists and dictionaries themselves, cannot be used because they can change, which would make the dictionary unreliable.

# Example 1: Lists as Keys (not allowed) :
# Trying to use a list as a key (will cause an error)
try:
    my_dict = {[1, 2, 3]: 'value'}
except TypeError as e:
    print(e)  # Output: unhashable type: 'list'


# Example 2: Dictionaries as Keys (not allowed) :
# Trying to use a dictionary as a key (will cause an error)
try:
    my_dict = {{'key': 'value'}: 'value'}
except TypeError as e:
    print(e)  # Output: unhashable type: 'dict'

# Summary : In simple terms, dictionary keys need to be immutable to ensure they stay the same and can be found quickly by the dictionary. Immutable types like strings and numbers work well as keys, while mutable types like lists and dictionaries do not work because they can change, making the dictionary unreliable.
```
